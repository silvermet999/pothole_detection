```python
import torch
%cd {HOME}
# Load the PyTorch model from the .pt file
model_path = f"{HOME}/runs/train/yolov5s_results_final/weights/best.pt"  # Replace with the actual path to your .pt file
loaded_model = torch.load(model_path, map_location=torch.device("cpu"))  # Load on CPU



# Remove training-related attributes
attributes_to_remove = ["best_fitness", "ema", "updates", "optimizer"]
for attribute in attributes_to_remove:
    if attribute in loaded_model:
        del loaded_model[attribute]

# Save the cleaned model (without training-related attributes) to a new checkpoint
cleaned_model_path = f"{HOME}/runs/train/yolov5s_results_final/weights/cleaned_model.pt"
torch.save(loaded_model, cleaned_model_path)


# Check for null values in the model's parameters

```

```python
cleaned_model = torch.load(cleaned_model_path)
has_null_values = False
for param_name, param_value in cleaned_model.items():
    if param_value is None:
        has_null_values = True
        print(f"Null value found in parameter: {param_name}")

if has_null_values:
    print("The model contains null values.")
else:
    print("The model does not contain null values.")
```
